# Clean Dataset

```python
print(hypo.isnull().sum())
```

```python
hypo.replace('?', np.nan, inplace=True)
```

```python
print(hypo.isnull().sum())
```

```python
print("Unique values in 'sex':", hypo['sex'].unique())
print("Unique values in 'Class':", hypo['Class'].unique())
```

```python
# Drop unnecessary columns
columns_to_drop = ['TSH_measured', 'T3_measured', 'TT4_measured', 'T4U_measured',
                   'FTI_measured', 'TBG_measured', 'TBG', 'referral_source']
hypo.drop(columns=columns_to_drop, inplace=True)
```

```python
# Convert categorical columns ('t' -> 1, 'f' -> 0)
binary_cols = ['on_thyroxine', 'query_on_thyroxine', 'on_antithyroid_medication',
               'sick', 'pregnant', 'thyroid_surgery', 'I131_treatment', 'query_hypothyroid',
               'query_hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych']
hypo[binary_cols] = hypo[binary_cols].applymap(lambda x: 1 if x == 't' else (0 if x == 'f' else np.nan))
```

```python
# Convert 'sex' column ('M' -> 1, 'F' -> 0)
hypo['sex'] = hypo['sex'].map({'M': 1, 'F': 0})
```

```python
# **Fix: Encode 'Class' as Multi-Class**
from sklearn.preprocessing import StandardScaler, LabelEncoder
label_encoder = LabelEncoder()
hypo['Class'] = label_encoder.fit_transform(hypo['Class'])
```

```python
print("Class Mapping:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```

```python
# Convert numerical columns to float
num_cols = ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']
hypo[num_cols] = hypo[num_cols].astype(float)
```

```python

num_cols_for_median = hypo.select_dtypes(include=['number']).columns  # Only numeric columns
hypo[num_cols_for_median] = hypo[num_cols_for_median].fillna(hypo[num_cols_for_median].median())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
hypo[num_cols] = scaler.fit_transform(hypo[num_cols])
```

```python
from sklearn.metrics import accuracy_score
```

```python
x = hypo.drop('Class', axis=1)
y = hypo['Class']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy_score', acc * 100)
```

```python
from sklearn.metrics import accuracy_score, classification_report # Import classification_report here

print("Classification_report")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy_score', acc * 100)
```

```python
print("Classification_report")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

dec = DecisionTreeClassifier()
dec.fit(x_train, y_train)
y_pred = dec.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy_score', acc * 100)
```

```python
print("Classification_report")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy_score', acc * 100)
```

```python
print("Classification_report")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

kne = KNeighborsClassifier()
kne.fit(x_train, y_train)
y_pred = kne.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy_score', acc * 100)
```

```python
print("Classification_report")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train, y_train)
y_pred = nb.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print('Accuracy_score', acc * 100)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate_model(model, model_name):
    # Get predictions
    y_pred = model.predict(x_test)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    # Classification Report
    print(f"\nClassification Report for {model_name}:\n")
    print(classification_report(y_test, y_pred))

    # Plot Confusion Matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()
```

```python
models = {
    "Logistic Regression": lr,
    "Support Vector Machine": svc,
    "Decision Tree": dec,
    "Random Forest": rf,
    "K-Nearest Neighbors": kne,
    "Naive Bayes": nb,
    "Gradient Boosting": gb,
    "XGBoost": xgb,
}

for model_name, model in models.items():
    evaluate_model(model, model_name)
```
